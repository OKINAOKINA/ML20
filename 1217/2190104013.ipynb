```
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

class function1():
    def __init__(self):
        """
        self.boundaries is the range of x and y axis
        """
        self.boundaries = np.array([-5.12, 5.12])

    def f(self, x, y):
        """
        Function (Z) value
        """
        t1 = 10 * 2
        t2 = x**2+y**2
        t3 = - 10 * np.cos(2 * np.pi * x)+10 * np.cos(2 * np.pi * y)
        return t1 + t2 + t3

class function2():
    def __init__(self):
        """
        self.boundaries is the range of x and y axis
        """
        self.boundaries = np.array([-5, 5])

    def f(self, x, y):
        """
        Function (Z) value
        """
        t1 = 100 * (y - x**2) ** 2
        t2 = (x - 1) ** 2
        return t1+t2

fig = plt.figure(figsize=(60, 60))
ax1 = fig.add_subplot(2, 1, 1, projection='3d')
ax2 = fig.add_subplot(2, 1, 2, projection='3d')

f1 = function1()

x1 = np.linspace(function1().boundaries[0], f1.boundaries[1], 4096)
y1 = np.linspace(f1.boundaries[0], f1.boundaries[1], 4096)
x1, y1 = np.meshgrid(x1, y1)
z1 = f1.f(x1,y1)

f2 = function2()

x2 = np.linspace(f2.boundaries[0], f2.boundaries[1], 4096)
y2 = np.linspace(f2.boundaries[0], f2.boundaries[1], 4096)
x2, y2 = np.meshgrid(x2, y2)
z2 = f2.f(x2,y2)

ax1.plot_surface(x1, y1, z1)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')

ax2.plot_surface(x2, y2, z2)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')

plt.show()
```
